#### 1. 선형 회귀모델

```python
from sklearn.linear_model import LinearRegression
```

```python
# 평가하는 함수를 만들면, 추후 다른 모델에도 바로 적용 가능!
def eval_regression(model):
    Y_train_pred = model.predict(X_train)
    Y_test_pred = model.predict(X_test)
    print('학습 능력')
    print('R Square : ', r2_score(Y_train, Y_train_pred))
    print('MSE : ', mean_squared_error(Y_train, Y_train_pred))
    print('일반화 능력')
    print('R Square : ', r2_score(Y_test, Y_test_pred))
    print('MSE : ', mean_squared_error(Y_test, Y_test_pred))
```

```python
# 평가
from sklearn.metrics import r2_score, mean_squared_error
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
## model 평가
eval_regression(model)
```

#### 2. 다항회귀분석

```python
X_train.columns
```

```python
## 다항회귀분석 실행
X_poly = PolynomialFeatures(degree=2).fit_transform(X_train) #degree=차수수
pd.DataFrame(X_poly)
```

```python
# 회귀계수 확인
model_linear['linearregression'].coef_ 
```

```python
# 다항회귀모델 평가
eval_regression(model_linear) 
```

#### 3. 규제선형모델

(1) Lasso

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
## 파이프라인으로 규제선형회귀 분석
model_pipe1 = make_pipeline(StandardScaler(),Lasso()) # 표준화 + Lasso 회귀 적용
## 스케일링 : Lasso 회귀처럼 규제를 사용하는 모델에서 중요 _ 특성들의 스케일이 다르면 규제가 잘못 적용될 수 있기 때문에 필수적으로 해야함.
model_pipe1
```

```python
hyperparam = {'lasso__alpha' : (0.5, 1, 1.5)} #클수록 강도가 강해짐짐
grid_model = GridSearchCV(model_pipe1, param_grid = hyperparam, cv=3, 
                          scoring='r2', n_jobs=-1)
grid_model.fit(X_train, Y_train)
best_model = grid_model.best_estimator_
best_model
```

```python
# 모델 평가
eval_regression(best_model)
```

```python
# 회귀계수 확인
best_model['lasso'].coef_ #0이 된 변수는 필요없음(age 변수) -> 규제 대상이 되어서 지워도 된다.
```

```python
df_coef = pd.DataFrame()
df_coef['Coef'] = best_model['lasso'].coef_
df_coef['X'] = X.columns
```

```python
px.bar(df_coef, x='X', y='Coef')
```

(2) Ridge

```python
from sklearn.linear_model import Ridge
```

```python
## 파이프라인으로 규제선형회귀 분석 - L2규제
model_pipe2 = make_pipeline(StandardScaler(), Ridge()) # 스케일링 + L2 규제
model_pipe2
```

```python
hyperparam = {'ridge__alpha' : (0.1, 1, 1.5)}
grid_model = GridSearchCV(model_pipe2, param_grid = hyperparam, cv=3, 
                          scoring='r2', n_jobs=-1)
grid_model.fit(X_train, Y_train)
best_model2 = grid_model.best_estimator_
best_model2
```

```python
# 모델 평가
eval_regression(best_model2)
```

```python
df_coef = pd.DataFrame()
df_coef['Coef'] = best_model2['ridge'].coef_
df_coef['X'] = X.columns
```

```python
px.bar(df_coef, x='X', y='Coef')
```

(3) Elastic Net

```python
from sklearn.linear_model import ElasticNet
```

```python
## 파이프라인으로 규제선형회귀 분석 - Elastic net
model_pipe3 = make_pipeline(StandardScaler(), ElasticNet()) # 스케일링 + elastic net
model_pipe3
```

```python
hyperparam = {'elasticnet__alpha' : (0.1, 1, 1.5)}
grid_model = GridSearchCV(model_pipe3, param_grid = hyperparam, cv=3, 
                          scoring='r2', n_jobs=-1)
grid_model.fit(X_train, Y_train)
best_model3 = grid_model.best_estimator_
best_model3
```

```python
# 모델 평가
eval_regression(best_model3)
```

```python
df_coef = pd.DataFrame()
df_coef['Coef'] = best_model3['elasticnet'].coef_
df_coef['X'] = X.columns
```

```python
px.bar(df_coef, x='X', y='Coef')
```
