## II. 지도 학습

#### 1) 데이터 핸들링

```python
# Outcome(y)값 확인
df['Outcome'].unique()
```

#### 사전 작업

#### Glucose, BloodPressure, SkinThickness,Insulin,BMI
0값 -> 중앙값으로 대체

```python
Glucose_median = df['Glucose'].median()
df['Glucose'] = df['Glucose'].replace(0, Glucose_median)

BloodPressure_median = df['BloodPressure'].median()
df['BloodPressure'] = df['BloodPressure'].replace(0, BloodPressure_median)

SkinThickness_median = df['SkinThickness'].median()
df['SkinThickness'] = df['SkinThickness'].replace(0, SkinThickness_median)

Insulin_median = df['Insulin'].median()
df['Insulin'] = df['Insulin'].replace(0, Insulin_median)

BMI_median = df['BMI'].median()
df['BMI'] = df['BMI'].replace(0, BMI_median)



```

#### 3) 학습데이터와 검증데이터를 분할

```python
# X 데이터의 학습, 검증 데이터 & Y 데이터의 학습, 검증데이터 분리
train_test_split(X,Y)
```

```python
X.describe()
```

####  4) 학습 수행 (Gradient Boosting 사용)



```python
from sklearn.ensemble import GradientBoostingClassifier 
```

```python
# 모델 선언
model = GradientBoostingClassifier() # 기본값: random_state=0 / max_depth=3 / learning_rate=0.1
# 모델 학습
model.fit(X_train, Y_train) #train만 들어가야함
```

```python
from sklearn.model_selection import GridSearchCV
```

#### 하이퍼파라미터 튜닝 

1. GridSearchCV 사용

```python

hyperparam = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0]
}

grid_model = GridSearchCV(
    model, 
    param_grid=hyperparam, 
    cv=3, #k=3 교차검증
    scoring='accuracy', 
    n_jobs=-1 #모델을 병렬처리 = -1
)

grid_model.fit(X_train, Y_train)
best_model = grid_model.best_estimator_
```

#### 5) 성능 확인


```python
from sklearn.metrics import accuracy_score
# 정확도 : 전체 데이터 중 얼만큼 정확히 분류했는지 비율
```

```python
# 훈련 데이터 예측값
Y_train_pred = model.predict(X_train) #학습에 사용한 데이터
# 테스트 데이터 예측값
Y_test_pred = model.predict(X_test) #학습에 사용하지 않은 새로운 데이터를 가지고 예측
```

```python
print('학습 정확도 : ', accuracy_score(Y_train, Y_train_pred)) # 여러 알고리즘에 대한 모델을 비교할 때 확인해야 함, 0.8정도가 좋음
print('검증 정확도 : ', accuracy_score(Y_test, Y_test_pred))
```

### 정리

#### 1. 분류 및 회귀 둘 다 가능한 모델 
#### 2. 가중치가 존재함 
#### 3. 잔차를 최소화하는 기법을 사용 
#### (잔차=실제값-예측값) 
#### 반복적으로 weak model/ 일반적으로  decision tree를 구축하여 잔차를 최소화하도록 하는 과정을 거친다. 


#### 3-1 손실함수

#### 손실 함수의 미분값이 0이 되는 지점을 잔차의 최솟값으로 판단한다. 즉 y=p인 지점을 발견해야한다.

#### 잔차 계산 예시


#### 이 표본에선 공부 시간과 성적의 관계성을 알아보고자 한다. 
#### 첫번째 예측 값은 표본의 평균값인 71.2로 설정한다. 그리고 결정나무를 만든다.  
#### 첫번째 표본 88은 88-71.2=16.8 이런 식으로 잔차를 계산한다. 
#### 하지만 의사나무에서 57 56/ 73 77처럼 두 개로 묶인 데이터들은 다른 방식으로 해야
#### 한다. 예를 들어 57 56은 각 잔차의 -14.2와 -15.2의 평균인 -14.7로 계산한다. 
#### 73 77 데이터도 마찬가지다. 
#### 그 후 원래 데이터값+가중치(0.1)*잔차를 계산하여 다음 값에 입력한다. 
#### Ex) 88+0.1*(16.8)=72.88, 57+0.1*(-14.7)=69.73 
#### 6개 표본을 다 계산하면 새로운 6개의 값들이 나오는데 그 값들을 또 앞선 방식 똑같이 
#### 해주면 된다. 이때 잔차는 기존 데이터값과 새로운 데이터값의 차이로 계산한다. 
#### Ex) (88-> 72.88의 잔차: 88-72.88=15.12)

#### 이 과정을 계속 거치다보면 잔차의 값들이 0으로 수렴하게 된다. 
#### (lteration 18, 19, 20  그래프 참고)
